```python
issued = df[df['Current Status'] == 'issued']
#issued.index
df.drop(issued.index, axis=0).head()
```

```python
#This drops all nans across all rows!
df.dropna()
```

```python
#This deletes nans bashed on threshold (more than 5 in this case)
df.dropna(thresh=5)
```

```python
df.dropna(subset=['Unit'])
```

```python
#create a list of all columns that we want to delete nans from
to_delete = ['Zipcode', 'Unit']
```

```python
#delete all nans from column list (all = nan in both , any = nan in any of the two)
df.dropna(subset = to_delete, how='any')
```

```python
df.replace(to_replace='expired',value='ended')
```

```python
#https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.Series.str.html
df[~df['Current Status'].str.match('issued')]
```

```python
# replace regex
#Here we save by redefining the df column (giving it by name)
df['Current Status'] = df['Current Status'].replace(to_replace='^w', value='W', regex=True)
```

```python
df.head()
```

```python
#Here we are saving again (but this time using inplace = True)
df['Current Status'].replace(to_replace='te$', value='ted', regex=True, inplace=True)
```
