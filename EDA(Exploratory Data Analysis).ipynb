```python
#Use map and the following dictionary to replace the genres.


dic = {
1:'Sports',
2:'Platform',
3:'Racing',
4:'Role-Playing',
5:'Puzzle',
6:'Misc',
7:'Shooter',
8:'Simulation',
9:'Action',
10:'Fighting',
11:'Adventure',
12:'Strategy'}
```

```python
def convert_gerne(x):
    return dic[x]

df['Genre'] = list(map(convert_gerne, df['Genre']))
df
```

```python
top_genres = df.groupby('Genre').count().sort_values(['Name'], ascending = False)
top_genres.head()
```
